```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from umap import UMAP
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv("adult.csv")  # Load your dataset here

# Data preprocessing
# Example:
# Encode categorical variables
data_encoded = pd.get_dummies(data)

# Scale numerical features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_encoded)

# Apply UMAP
umap = UMAP(n_components=2)
reduced_data = umap.fit_transform(scaled_data)

# Visualization
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=data['income'].map({'<=50K': 'blue', '>50K': 'deeppink'}))
plt.title("UMAP Visualization of Adult Income Data")
plt.xlabel("UMAP Dimension 1")
plt.ylabel("UMAP Dimension 2")
plt.show()
```
